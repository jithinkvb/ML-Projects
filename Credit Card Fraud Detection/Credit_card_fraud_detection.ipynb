This dataset is highly unbalanced

0 --> Legit Transactions

1 --> Fradulent Transaction

```python
# Seperating the data for analysis as legit and fraud
legit = cdata[cdata.Class == 0]
fraud = cdata[cdata.Class == 1]
```

Under sampling methode - We use this methode because our dataset is higly unbalanced so we nee dto balance the both type of data that we have here. therefore we creating a sample dataset that containing similer distribution of both fraud and legit data.- And here we take a random 514 similer sample from the legit transactions and we creat a sample dataset then we have 514 sample of both type of transactions

Here we build a sample dataset containing similer distribution of legit and fradulent transactions

number of fradulent transactions --> 514

```python
legit_sample = legit.sample(n = 514)
```

Now we have the 514 legit legit transactions. Then we concatinating the both type of transactions

```python
# concatination of both dataset
new_cdata = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_cdata.head()
```

```python
new_cdata.tail()
```

```python
new_cdata['Class'].value_counts()
```

```python
new_cdata.groupby('Class').mean()
```

Splitting the dtaset into features and target

```python
x = new_cdata.drop(['Class'],axis = 1)
```

```python
y = new_cdata['Class']
```

```python
x
```

```python
y
```

```python
# Splitting the dataset into training and testing
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

Here we using stratify= y because o get the evenly ditribution of both classes from xtrain and xtest

# Model Training

Here our present problem is a binary classification type problem therfor here we using logistic regression to train our model

```python
model = LogisticRegression()
```

```python
# training the logistic regression model with the training data
model.fit(x_train, y_train)
```

Model evaluation

Accuracy score

```python
# Accuracy on training data
x_pred = model.predict(x_train)
training_data_acc = accuracy_score(x_pred, y_train)
```

```python
print('Training data Accuracy :', training_data_acc)
```

```python
# accuracy on testing data
x_test_prediction = model.predict(x_test)
test_data_acc = accuracy_score(x_test_prediction, y_test)
```

```python
print('Test data accuracy : ', test_data_acc)
```
